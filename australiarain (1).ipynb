```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df=pd.read_csv("C:\\Users\\anjal\\OneDrive\\Desktop\\ml assignments sem5\\weatherAUS.csv")
df=df.dropna(subset=["MinTemp","Rainfall"])
df
```

```python
correlation_matrix = df[['MinTemp', 'Rainfall']].corr()
print("Correlation Matrix:")
print(correlation_matrix)
```

```python
X=df["MinTemp"].values.reshape(-1,1) #independent variable
y=df["Rainfall"].values.reshape(-1,1) #dependent variable 
```

```python
model=LinearRegression()
model.fit(X,y)
prediction=model.predict(X)
```

```python
mse = mean_squared_error(y, prediction)
r2 = r2_score(y, prediction)

# Print the results
print(f"Mean Squared Error (MSE): {mse}")
print(f"R^2 Score: {r2}")
```

```python
plt.scatter(X,y,color="blue",label="Actual data")
plt.plot(X,prediction,color="red",linewidth=2,label="Regression Line")
plt.xlabel("MinTemp")
plt.ylabel("Rainfall")
plt.title("Linear Regression on the Rain in Australia dataset")
plt.legend()
plt.grid(True)
plt.show()
```
